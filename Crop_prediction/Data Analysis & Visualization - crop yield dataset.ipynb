```python
# importing necessary libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset

crop_data=pd.read_csv("crop_production.csv")
crop_data
```

<b> Exploratory Data Analysis and Data Pre-processing </b> 

```python
crop_data.shape

#rows X columns
```

```python
# dataset columns
crop_data.columns
```

```python
# statistical inference of the dataset

crop_data.describe()
```

```python
# viewing unique crops available in the dataset

print(crop_data['Crop'].unique())
```

```python
crop_data.info()
```

```python
# Checking missing values of the dataset in each column
crop_data.isnull().sum()
```

```python
# Dropping missing values 
crop_data = crop_data.dropna()
crop_data
```

```python
#checking
crop_data.isnull().values.any()
```

```python
# Displaying State Names present in the dataset
crop_data.State_Name.unique()
```

```python
# Adding a new column Yield which indicates Production per unit Area. 

crop_data['Yield'] = (crop_data['Production'] / crop_data['Area'])
crop_data.head(10) 
```

<i> Visualizing the dataset : </i>

```python

sorted_data = crop_data.sort_values(by="Production", ascending=False)

# Plot with a colorful palette
sns.barplot(x="State_Name", y="Production", data=sorted_data, palette="viridis")

# Rotate x-axis labels for readability
plt.xticks(rotation=90)
```

```python
# Sort data by Yield (optional but helps readability)
sorted_data = crop_data.sort_values(by="Yield", ascending=False)

# Create the barplot with a colorful palette
sns.barplot(x="State_Name", y="Yield", data=sorted_data, palette="Set2")

# Rotate x-axis labels
plt.xticks(rotation=90)

# Add a title and improve layout
plt.title("Yield by State")
plt.tight_layout()
plt.show()
```

```python
# Sort data by Crop_Year (just in case)
sorted_data = crop_data.sort_values(by="Crop_Year")

# Create the barplot with a colorful palette
sns.barplot(x="Crop_Year", y="Production", data=sorted_data, palette="cubehelix")

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Add a title and improve layout
plt.title("Crop Production Over the Years")
plt.tight_layout()
plt.show()
```

```python
# Sort data by Crop_Year
sorted_data = crop_data.sort_values(by="Crop_Year")

# Plot Yield over the years with a color palette
sns.barplot(x="Crop_Year", y="Yield", data=sorted_data, palette="mako")

# Rotate x-axis labels
plt.xticks(rotation=90)

# Add a title and improve layout
plt.title("Yield Over Crop Years")
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x="Season", y="Yield", data=crop_data, marker="o")

plt.xticks(rotation=90)
plt.title("Yield by Season")
plt.tight_layout()
plt.show()
```

```python

# Create the line plot for Production over Season
sns.lineplot(x="Season", y="Production", data=crop_data, marker="o")

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Add a title and improve layout
plt.title("Production by Season")
plt.tight_layout()
plt.show()
```

```python
# Create the bar plot for Yield by Crop
sns.barplot(x="Crop", y="Yield", data=crop_data, palette="coolwarm")

# Rotate x-axis labels for better readability
plt.xticks(rotation=60)

# Add a title and improve layout
plt.title("Yield by Crop")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot with regression line (faster for large datasets)
sns.regplot(x="Area", y="Production", data=crop_data)

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.title("Production vs Area (with Regression Line)")
plt.tight_layout()
plt.show()
```

```python


# Use a subset of the data to speed up the plotting
subset_data = crop_data.sample(1000)  # Adjust the number of rows for performance

# Create the jointplot with a regression line
joint = sns.jointplot(x="Area", y="Production", data=subset_data, kind="reg")

# Rotate x-axis labels using the axis object of the jointplot
joint.ax_joint.set_xticklabels(joint.ax_joint.get_xticklabels(), rotation=90)

# Display the plot
plt.show()
```

Conclusions Obtained from the dataset:

1. Kerela is the largest producer of crop in India.
2. Production per unit area (Yield ) of Puducherry is maximum.
3. In the year 2011, crop yield was maximum.
